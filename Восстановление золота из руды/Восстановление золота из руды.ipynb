## Введение

### Описание проекта

`Восстановление золота из руды` - проект, целью которого является построение модели для предсказания коэффициента восстановления золота из золотосодержащей руды. Модель необходима для оптимизации производства, чтобы не запускать предприятие с убыточными характеристиками.

Заказчиком является «Цифра» - компания разрабатывающая решения для эффективной работы промышленных предприятий.

**Технологический процесс обработки золотосодержащей руды:**



**Этап 1 -  Первичная обработка**
* `Входное сырье` - добытая золотосодержащая руда;
* `Выходной продукт` - дроблёная смесь золотосодержащей руды;
* `Комментарий:` Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку;

**Этап 2 - Флотация ('rougher')**
* `Входное сырье` - дроблёная смесь золотосодержащей руды;
* `Выходной продукт` - черновой концентрат и «отвальные хвосты этапа флотации»;
* `Комментарий:` Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.

**Этап 3 и 4 - Очистка ('primary_cleaner', 'secondary_cleaner')**
* `Входное сырье` - черновой концентрат и «отвальные хвосты этапа флотации»;
* `Выходной продукт` - финальный концентрат и «отвальные хвосты этапа очистки 1 и 2»;
* `Комментарий:` Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Расчет эффективности обогащения

* Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery.

```python
# напишем функцию, которая вычислит эффективность обогащения на обучающей выборке для сравнения с `rougher.output.recovery`
# формула эффективности обогащения:
# Recovery = ( C*(F - T) ) / ( F*(C - T) )
# где С - rougher.output.concentrate_ag  (финальный концентрат золота)
#     F - rougher.input.feed_au          доля золота в сырье/концентрате до флотации/очистки (черновой концентрат золота)
#     T - rougher.output.tail_au         доля золота в отвальных хвостах после флотации/очистки (отвальный концентрат золота).

def output_recovery(concentrate, feed, tail):
    
    recovery =  ( concentrate*(feed - tail) ) / ( feed*(concentrate - tail) )
    
    return recovery*100 # домножим на 100, т.к. величина в датафрейме в %
    
# вызовем функцию    
train['rougher.output.recovery_calc'] = output_recovery(train['rougher.output.concentrate_au'],
                                                           train['rougher.input.feed_au'],
                                                           train['rougher.output.tail_au'])
```

* Найдите MAE между вашими расчётами и значением признака. Опишите выводы.

```python
# с помощью функции mean_absolute_error вычислим MAE между 'recovery_calc' и столбцом 'rougher.output.recovery'
rough_mae = mean_absolute_error(train['rougher.output.recovery'],
                                train['rougher.output.recovery_calc'])
rough_mae
```

**Промежуточный вывод**
* Была написана функция, которая вычислила эффективность обогащения на обучающей выборке, для проверки показателя `rougher.output.recovery`;
* Найденное `MAE` (mean absolute error) или средняя абсолютная погрешность между расчётом и значением признака 'rougher.output.recovery' составило `9.73512347450521e-15`. Данный результат говорит нам о высокой степени точности и корректности расчета значения признака в столбце 'rougher.output.recovery'.

## Модель

### Функция вычисления итоговой sMAPE

* Напишем функцию для вычисления метрики качества sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»), руководствуясь формулой из условий задачи:

```python
def symmetricMAPE (target, prediction):
    
    smape = sum( abs(target - prediction)/( (abs(target) + abs(prediction))/2 ) ) / len(target) * 100
    
    return smape 
```

#### Модель дерева решений

```python
%%time
# результаты итоговых вычислений запишем в словарь списков
dictionary = {'tree_depth':list(),
              'smape':list()}


# разобьем train-выборку на переменные features и target для процессов rougher и final
target_r   = train_r['rougher.output.recovery']
features_r = train_r.drop('rougher.output.recovery', axis=1)

target_f   = train_f['final.output.recovery']
features_f = train_f.drop('final.output.recovery', axis=1)

# присвоим объекту scale структуру данных StandardScaler
scaler = StandardScaler()

# с помощью 'fit_transform' выполним масштабирование признаков на обучающей выборке
features_r = pd.DataFrame(scaler.fit_transform(features_r), columns=rougher_columns[:22])
features_f = pd.DataFrame(scaler.fit_transform(features_f), columns=final_columns[:30])

# циклом переберем гиперпараметр 'max_depth' и подберем наилучшую модель дерева решений
for depth in range(1, 30, 1):
    
    # создадим объект 'model' структуры данных DecisionTreeRegressor
    # зададим псевдослучайность random_state=12345
    model = DecisionTreeRegressor(random_state=12345,
                                  max_depth=depth)

    # на каждом шаге перебора глубины результаты промежуточных sMAPE-вычислений запишем в список
    smape   = []
    
    
    # --------------------------------------------
    # БЛОК 1: разбивка TRAIN-выборки на обучающую и валидационную, запуск кросс-валидации
    # --------------------------------------------

    # 1. зададим размер блока, если их всего 4 (соттношение обучающей и валидационной выборок 3:1)
    sample_size = int(len(features_r)/4)

    # 2. циклом разобьем обучающий и валидационный набор данных выборки train
    for i in range(0, len(features_r), sample_size):
    
        # массив из индексов для валидационного блока
        valid_indexes = list(range(i, i+sample_size))
            
        # массив из индексов для обучающей выборки
        train_indexes = list(range(i)) + list(range(i+sample_size, len(features_r)))
            
        # 3. разобьем переменные features и target на выборки features_train, target_train, features_valid, target_valid
        
        # 3.1 для процессов rougher
        features_train_r  = features_r.iloc[train_indexes]
        target_train_r    = target_r.iloc[train_indexes]
        features_valid_r  = features_r.iloc[valid_indexes]
        target_valid_r    = target_r.iloc[valid_indexes]

        # 3.2 для процессов final
        features_train_f  = features_f.iloc[train_indexes]
        target_train_f    = target_f.iloc[train_indexes]
        features_valid_f  = features_f.iloc[valid_indexes]
        target_valid_f    = target_f.iloc[valid_indexes]
    
       
        # --------------------------------------------
        # БЛОК 2: расчет sMAPE-метрики для этапа ROUGHER
        # --------------------------------------------
    
        # обучим модель на ОБУЧАЮЩЕЙ выборке
        model.fit(features_train_r, target_train_r)

        # произведем предсказания на ВАЛИДАЦИОННОЙ выборке
        predictions = model.predict(features_valid_r)

        # рассчитаем качество предсказаний на ОБУЧАЮЩЕЙ выборке с помощью sMAPE-метрики для этапа ROUGHER  
        smape_r = symmetricMAPE(target_valid_r, predictions)
    

        # --------------------------------------------
        # БЛОК 3: расчет sMAPE-метрики для этапа FINAL
        # --------------------------------------------
    
        # обучим модель на обучающей выборке
        model.fit(features_train_f, target_train_f)

        # произведем предсказания на ОБУЧАЮЩЕЙ выборке
        predictions = model.predict(features_valid_f)

        # рассчитаем качество предсказаний на ОБУЧАЮЩЕЙ выборке с помощью sMAPE-метрики для этапа FINAL
        smape_f = symmetricMAPE(target_valid_f, predictions)

    
        # --------------------------------------------
        # БЛОК 4: итоговый расчет sMAPE-метрики
        # --------------------------------------------

        smape_calc = 0.25 * smape_r + 0.75 * smape_f
        
        # добавим рассчитанное значение в список smape
        smape.append(smape_calc)
        
    # --------------------------------------------
    # БЛОК 5: Запись в словарь
    # --------------------------------------------
    
    # запишем СРЕДНЕЕ из полученных значения sMAPE-метрики в словарь
    dictionary['smape'].append(pd.Series(smape).mean())
        
    # запишем гиперпараметры дерева
    dictionary['tree_depth'].append(depth)

    
# --------------------------------------------
# БЛОК 6: Вывод результатов
# --------------------------------------------

data = pd.DataFrame(dictionary)
print(data)
print()
print('Параметры наилучшей модели дерева решений:')
print(data[data['smape'] == data['smape'].min()])
```

```python
%%time
# результаты итоговых вычислений запишем в словарь списков
dictionary = {'smape':list()}


# разобьем train-выборку на переменные features и target для процессов rougher и final
target_r   = train_r['rougher.output.recovery']
features_r = train_r.drop('rougher.output.recovery', axis=1)

target_f   = train_f['final.output.recovery']
features_f = train_f.drop('final.output.recovery', axis=1)

# присвоим объекту scale структуру данных StandardScaler
scaler = StandardScaler()

# с помощью 'fit_transform' выполним масштабирование признаков на обучающей выборке
features_r = pd.DataFrame(scaler.fit_transform(features_r), columns=rougher_columns[:22])
features_f = pd.DataFrame(scaler.fit_transform(features_f), columns=final_columns[:30])

    
# создадим объект 'model' структуры данных LinearRegression

model = LinearRegression()

# на каждом шаге перебора глубины результаты промежуточных sMAPE-вычислений запишем в список
smape   = []
    
    
# --------------------------------------------
# БЛОК 1: разбивка TRAIN-выборки на обучающую и валидационную, запуск кросс-валидации
# --------------------------------------------

# 1. зададим размер блока, если их всего 4 (соттношение обучающей и валидационной выборок 3:1)
sample_size = int(len(features_r)/4)

# 2. циклом разобьем обучающий и валидационный набор данных выборки train
for i in range(0, len(features_r), sample_size):
    
    # массив из индексов для валидационного блока
    valid_indexes = list(range(i, i+sample_size))
            
    # массив из индексов для обучающей выборки
    train_indexes = list(range(i)) + list(range(i+sample_size, len(features_r)))
            
    # 3. разобьем переменные features и target на выборки features_train, target_train, features_valid, target_valid
        
    # 3.1 для процессов rougher
    features_train_r  = features_r.iloc[train_indexes]
    target_train_r    = target_r.iloc[train_indexes]
    features_valid_r  = features_r.iloc[valid_indexes]
    target_valid_r    = target_r.iloc[valid_indexes]

    # 3.2 для процессов final
    features_train_f  = features_f.iloc[train_indexes]
    target_train_f    = target_f.iloc[train_indexes]
    features_valid_f  = features_f.iloc[valid_indexes]
    target_valid_f    = target_f.iloc[valid_indexes]
    
       
    # --------------------------------------------
    # БЛОК 2: расчет sMAPE-метрики для этапа ROUGHER
    # --------------------------------------------
    
    # обучим модель на ОБУЧАЮЩЕЙ выборке
    model.fit(features_train_r, target_train_r)

    # произведем предсказания на ВАЛИДАЦИОННОЙ выборке
    predictions = model.predict(features_valid_r)

    # рассчитаем качество предсказаний на ОБУЧАЮЩЕЙ выборке с помощью sMAPE-метрики для этапа ROUGHER  
    smape_r = symmetricMAPE(target_valid_r, predictions)
    

    # --------------------------------------------
    # БЛОК 3: расчет sMAPE-метрики для этапа FINAL
    # --------------------------------------------
    
    # обучим модель на обучающей выборке
    model.fit(features_train_f, target_train_f)

    # произведем предсказания на ОБУЧАЮЩЕЙ выборке
    predictions = model.predict(features_valid_f)

    # рассчитаем качество предсказаний на ОБУЧАЮЩЕЙ выборке с помощью sMAPE-метрики для этапа FINAL
    smape_f = symmetricMAPE(target_valid_f, predictions)

    
    # --------------------------------------------
    # БЛОК 4: итоговый расчет sMAPE-метрики
    # --------------------------------------------

    smape_calc = 0.25 * smape_r + 0.75 * smape_f
        
    # добавим рассчитанное значение в список smape
    smape.append(smape_calc)
        
# --------------------------------------------
# БЛОК 5: Запись в словарь
# --------------------------------------------
    
# запишем СРЕДНЕЕ из полученных значения sMAPE-метрики в словарь
dictionary['smape'].append(pd.Series(smape).mean())

    
# --------------------------------------------
# БЛОК 6: Вывод результатов
# --------------------------------------------

data = pd.DataFrame(dictionary)
print('Параметры модели линейной регрессии:')
print(data[data['smape'] == data['smape'].min()])
```

2. Обучены разные модели (дерева решений, случайного леса, линейной регрессии) и оценено их качество кросс-валидацией. Результаты показали, что:
