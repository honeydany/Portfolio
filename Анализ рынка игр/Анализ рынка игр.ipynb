## Введение

**Описание проекта**

Интернет-магазин «Стримчик» занимается продажей по всему миру компьютерных игр. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Исходные данные:** 

* Перед нами исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation) до 2016 года.
* В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Описание данных:** 

* `Name` — название игры
* `Platform` — платформа
* `Year_of_Release` — год выпуска
* `Genre` — жанр игры
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
* `EU_sales` — продажи в Европе (миллионы проданных копий)
* `JP_sales` — продажи в Японии (миллионы проданных копий)
* `Other_sales` — продажи в других странах (миллионы проданных копий)
* `Critic_Score` — оценка критиков (максимум 100)
* `User_Score` — оценка пользователей (максимум 10)
* `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Подготовка данных

### Исправление регистра заголовков

```python
# получим название всех колонок
print(df.columns)

# заменим названия
df.columns = df.columns.str.lower()
df.columns
```

### Исправление типа данных

При выполнении кода `df = df.astype({'year_of_release':'int64', 'user_score':'float64'})` для замены типа данных были получены следующие ошибки:
* В столбце `user_score` получена ошибка: `ValueError: could not convert string to float: tbd`.
    
    Сокращение `tbd` часто используется в сфере компьютерных игр и расшифровывается очень просто - `To Be Determined`, то есть "Будет определено". Стоит отметить также, что у этой аббревиатуры могут быть и иные значения, например `To Be Done` - должно быть сделано, или `To Be Discussed` - подлежит обсуждению. Иными словами `tbd` обозначает пропуски, которые возникли в связи с отсутствием данных, т.к. оценка игры на момент выгрузки не была произведена либо не внесена на платформу. Предлагается заменить их на `NaN`.


* В столбце `year_of_release`: `IntCastingNaNError: Cannot convert non-finite values (NA or inf) to integer`. 

    Что значит: `Ошибка Int-NaN: Не удается преобразовать конечные значения (NA или inf) в целое число`. Так как причину отсутствия значений в столбце `year_of_release` выяснить и заполнить корректными значениями пока не представляется возможным, предлагаем удалить пропуски.


```python
# заменим 'tbd' на NaN в солбце 'user_score'
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan

# удалим отсутствующие значения в столбце 'year_of_release'
df.dropna(subset=['year_of_release'], inplace=True)

# повторим замену типа данных в столбцах 'year_of_release', 'user_score'
df = df.astype({'year_of_release':'int64', 'user_score':'float64'})

# проверим
df.dtypes
```

### Выпуск игр по годам

* Посмотрите, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

```python
# сделаем копию нашего датафрейма и в дальнейшем будем работать с ней
data = df.copy()

# сгруппируем данные по годам, подсчитаем кол-во игр
a = pd.pivot_table(data, index='year_of_release', values='name', aggfunc='count').reset_index()
a['growth,%'] = a['name'].pct_change().round(4)*100
a
```

```python
a.plot(kind='bar',
       x='year_of_release',
       y='growth,%',
       legend=True,
       figsize=(20,10))
plt.title('Прирост числа игр г/г,%')
plt.show()
```

```python
a.plot(kind='bar',
       x='year_of_release',
       y='name',
       legend=True,
       figsize=(20,10))
plt.title('Выпуск игр по годам, шт')
plt.show()
```

```python
# Предлагаем убрать данные по играм, выпущенным до 1991 года, т.к. в игровой индустрии отсутствовал устойчивый выпуск контента
data = data.query('year_of_release > 1990')
```

### Самые популярные платформы (топ-5)

```python
# отфильтруем данные за актуальный период с 2013 по 2016 гг
data = data[data['year_of_release'].isin(years)]

# напишем цикл для построения сводных таблиц и графиков по требуемым регионам в разрезе платформ
regions = ['na_sales', 'eu_sales','jp_sales']
title_name = ['Самые популярные платформы (топ-5) в Северной Америке',
              'Самые популярные платформы (топ-5) в Европе',
              'Самые популярные платформы (топ-5) в Японии']

for index in range(len(regions)):
    
    # сводная по платформам
    top_platforms = pd.pivot_table(data=data,
                               index='platform',
                               values=regions[index],
                               aggfunc='sum',
                               fill_value=0)
    
    # фильтр по убыванию
    top_platforms.sort_values(by=regions[index],
                              ascending=False,
                              inplace=True)
    # расчет доли продаж
    top_platforms['share'] = (top_platforms/top_platforms.sum()).round(3)
    
    # фильр первых 5 строк
    display(top_platforms.head(5))
    
    # вывод стобчатой диаграммы
    top_platforms[regions[index]].head(5).plot(kind='barh',
                                               title=title_name[index],
                                               figsize=(20,10))
    plt.show()
    
    # вывод круговой диаграммы
    top_platforms['share'].plot(kind='pie',
                                title=title_name[index],
                                figsize=(20,10))
    
    plt.show()
```

### Влияние рейтинга ESRB на продажи в регионах

```python
# напишем цикл для построения сводных таблиц и графиков по требуемым регионам в разрезе оценок рейтинга ESRB
title_name = ['Продажи игр в Северной Америке в зависимости от оценок рейтинга ESRB',
              'Продажи игр в Европе в зависимости от оценок рейтинга ESRB',
              'Продажи игр в Японии в зависимости от оценок рейтинга ESRB']

# предварительно позиции без оценок ESRB заменим на 'No_rating'
data['rating'].fillna(value='No_rating', inplace=True)

for index in range(len(regions)):
    
    # сводная по оценкам рейтинга ESRB
    ESRB_rating = pd.pivot_table(data=data,
                               index='rating',
                               values=regions[index],
                               aggfunc='sum',
                               fill_value=0)
 
    # фильтр по убыванию
    ESRB_rating.sort_values(by=regions[index],
                          ascending=False,
                          inplace=True)
     
    # фильр первых 5 строк
    print(ESRB_rating.head(5))
    ESRB_rating.head(5).plot(kind='barh',
                           title=title_name[index],
                           figsize=(20,10))
    plt.show()
```

* Однако однозначно утверждать о наличии связи нельзя, т.к., например, мы ничего достоверно не знаем о возрастных категориях людей, которые эти игры покупали;

## Проверка гипотез

* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Для проверки гипотезы "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые" сформулируем две гипотезы:

* `Нулевая гипотеза (H_0)`: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"
* `Альтернативная двусторонняя гипотеза (H_1)`: "Средние пользовательские рейтинги платформ Xbox One и PC отличаются".

Основным инструментом проверки гипотез о равенстве средних двух генеральных совокупностей, как правило, является двусторонний t-test Стьюдента. Прежде, чем его использовать, необходимо проверить несколько условий его применимости, а именно:
* генеральные совокупности не должны зависеть друг от друга;
* выборочные средние должны быть нормально распределены;
* дисперсии рассматриваемых генеральных совокупностей должны быть равны.

Для проверки условий применимости выполним необходимые расчеты:

```python
# отфильтруем выборки для платформ Xbox One и PC
data_xone = data.query('platform == "XOne"')
data_pc = data.query('platform == "PC"')

# построим гистограммы распределения пользовательских рейтингов платформ
ax = data_xone['user_score'].plot(kind='hist',
                                  legend=True,
                                  label='XOne')

data_pc['user_score'].plot(kind='hist',
                           ax=ax,
                           grid=True,
                           legend=True,
                           histtype='step',
                           linewidth=2,
                           label='PC',
                           title='Распределение пользовательских рейтингов игровых платформ')
plt.show()

# рассчитаем дисперсии пользовательских рейтингов платформ
print('Дисперсия пользовательских рейтингов платформы "XOne" - ', round(np.var(data_xone['user_score']), 3))
print('Дисперсия пользовательских рейтингов платформы "PC" - ', round(np.var(data_pc['user_score']), 3))
```

На основе проведенных расчетов можно сделать вывод, что `не все` условия применимости t-test Стьюдента соблюдены:
* генеральные совокупности не зависят друг от друга - `условие выполнено`;
* выборочные средние нормально распределены (см. график) - `условие выполнено`;
* дисперсии рассматриваемых генеральных совокупностей равны - `условие не выполнено`.

Поэтому при выполнении двустороннего t-test Стьюдента укажем параметр `equal_var = False`, то есть тест посчитает дисперсии выборок по отдельности. Выполним двусторонний t-test Стьюдента:

```python
# для целей проверки гипотезы исключим пропуски в данных:
data_xone = data_xone[data_xone['user_score'].notna()]
data_pc = data_pc[data_pc['user_score'].notna()]

# зададим критический уровень статистической значимости 5% - стандартно :))
alpha = .05 # если p-value окажется меньше него - отвегнем гипотезу
    
results = st.ttest_ind(data_xone['user_score'],
                       data_pc['user_score'],
                       equal_var=False) 
# по умолчанию параметр equal_var=True, то есть тест 
# учитывает, что дисперсии выборок одинаковы

# p-value - кф, характеризующий вероятность получить наблюдаемый результат при условии,
# что нулевая гипотеза Н_0 верна
print('p-значение: ', results.pvalue)
print('Статистика разности: ', results.statistic)

if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Убедимся в непротиворечивости полученных результатов:

```python
print('Средняя оценка пользователей "XOne" - ', data_xone['user_score'].mean())
print('Средняя оценка пользователей "PC" - ', data_pc['user_score'].mean())
```

**Вывод:**

Проведя t-test Стьюдента мы получили `p-значение:  0.14759594013430463`, что характеризует относительно высокую вероятность получить наблюдаемый результат при условии, что нулевая гипотеза Н_0 верна. `Статистика разности:  1.45192815744488` статистически не значима.

На критическом уровене статистической значимости 5% мы `не отвергаем нулевую гипотезу Н_0 "О равенстве средних пользовательских рейтингов платформ Xbox One и PC"`.
