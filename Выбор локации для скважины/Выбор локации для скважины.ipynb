## Загрузка и подготовка данных

```python
# импортируем необходимые библиотеки
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# импортируем данные
df0 = pd.read_csv('/datasets/geo_data_0.csv')
df1 = pd.read_csv('/datasets/geo_data_1.csv')
df2 = pd.read_csv('/datasets/geo_data_2.csv')

df = [df0, df1, df2]

# добавим в каждую таблицу признак региона
for i in range(0,3):
    df[i]['region'] = i

# соединим таблицы
df = pd.concat([df0, df1, df2])

# выведем основную информацию по датафреймам на экран
display(df)
df.info()
display(df.describe())
```

## Обучение и проверка модели

Обучите и проверьте модель для каждого региона:
* Разбейте данные на обучающую и валидационную выборки в соотношении 75:25.
* Обучите модель и сделайте предсказания на валидационной выборке.
* Сохраните предсказания и правильные ответы на валидационной выборке.
* Напечатайте на экране средний запас предсказанного сырья и RMSE модели.
* Проанализируйте результаты.

```python
# создадим словарь, в который запишем предсказания по регионам
pred_dict = {0:[],
             1:[],
             2:[]}


# циклом пройдем по датафреймам каждого региона
for i in range(3):
    
    region = df[df['region'] == i]
    # разделим признаки и целевой признак
    features = region.drop('product', axis=1)
    target = region['product']
    
    # разделим признаки и целевой признак на обучающий и валидационный набор
    features_train, features_valid, target_train, target_valid = train_test_split(features,
                                                                                  target,
                                                                                  test_size=0.25,
                                                                                  random_state=12345)
    # объекту model присвоим структуру данных LinearRegression
    model = LinearRegression()
    
    # обучим модель на обучающей выборке
    model.fit(features_train, target_train)
    
    # выполним предсказания на валидационной выборке
    predictions = model.predict(features_valid)
    
    # переведем список в структуру данных Series
    predictions = pd.Series(predictions,
                            index=features_valid.index)
    # запишем в словарь
    pred_dict[i] = predictions
    
    # рассчитаем rmse модели
    mse = mean_squared_error(pd.Series(target_valid), predictions)
    rmse = mse ** 0.5
    
    print(f'Средний объём запасов предсказанного сырья в 1 скважине для региона {i} - {predictions.mean():.2f} тыс. барр.')
    print(f'RMSE модели - {rmse:0.2f}')
    print()    
```

* рассчитана метрика качества `RMSE` для модели линейной регрессии.

    * `RMSE` модели - `0.89`
    
    2.3 Регион 3:

3. Наиболее точно модель линейной регрессии предсказала значения сырья в скважинах для региона №1, т.к. значение RMSE модели составило 0.89 тыс. барр. Предсказания по модели для регионов 0 и 2 показали результаты хуже - RMSE 37.58 и 40.03 тыс. барр. соответственно, что составляет практически половину от среднего запаса сырья в скважине этих регионов.

## Подготовка к расчёту прибыли

* Все ключевые значения для расчётов сохраните в отдельных переменных.
* Рассчитайте достаточный объём сырья для безубыточной разработки новой скважины. Сравните полученный объём сырья со средним запасом в каждом регионе.
* Напишите выводы по этапу подготовки расчёта прибыли.

```python
# Бюджет на разработку скважин в регионе — 10 млрд рублей.
Budget = 10 ** 10
# При нынешних ценах один баррель сырья приносит 450 рублей дохода.
# При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
Count = 200

Revenue_per_barrel = 450

# рассчитаем необходимый объем баррелей для достижения безубыточной добычи
breakeven_point = Budget / Count / 450 / 1000
print(f'Минимальный объем добычи сырья для достижения точки безубыточности 1 скважины - {breakeven_point:.2f} тыс. барр./скважину')
```

* Точка безубыточной добычи 1 скважины выше, чем средний объем добычи на 1 скважине в рассматриваемых регионах, следовательно, предварительно можно предположить, что добыча в рассматриваемых регионах может быть убыточна.

## Расчет прибыли и рисков 

### Расчёт прибыли по выбранным скважинам

Напишите функцию для расчёта прибыли по выбранным скважинам и предсказаниям модели:
* Выберите скважины с максимальными значениями предсказаний.
* Просуммируйте целевое значение объёма сырья, соответствующее этим предсказаниям.
* Рассчитайте прибыль для полученного объёма сырья.

```python
# напишем функцию для расчета прибыли
def profit(data, predictions, Budget, rev_per_barrel):
    
    # Выберем скважины с максимальными значениями предсказаний.
    predictions_sorted = predictions.sort_values(ascending=False)[:200]

    # Просуммируйте целевое значение объёма сырья (колонка 'production'), 
    # соответствующее предсказаниям по ТОП-200 скважинам
    total_recovery = data['product'][predictions_sorted.index].sum()
        
    # Рассчитайте прибыль для полученного объёма сырья.
    net_profit = (total_recovery * rev_per_barrel * 1000 - Budget) / 10**9
    
    return net_profit, total_recovery

# Циклом пройдем по предсказанным данным добычи каждого региона
for i in range(3):
    net_profit, total_recovery = profit(df[df['region'] == i], # отфильтруем данные по каждому региону
                                        pred_dict[i],          # из pred_dict возьмем предсказания по каждому региону
                                        Budget,                # общий бюджет на разработку 200 скважин
                                        450)                   # выручка при продаже барреля нефти
    
    
    # Выведем информацию на экран
    print(f'Общий объем добычи по 200 скважинам в регионе {i} - {total_recovery:.2f} тыс. барр., чистая прибыль - {net_profit:.2f} млрд. руб.')
```

### Учет рисков

* Примените технику Bootstrap с 1000 выборок, чтобы найти распределение прибыли.
* Найдите среднюю прибыль, 95%-й доверительный интервал и риск убытков. Убыток — это отрицательная прибыль.
* Напишите выводы: предложите регион для разработки скважин и обоснуйте выбор.

Для применения техники Bootstrap с 1000 выборок выполним следующее:
1. Предварительная подготовка;
2. Напишем два цикла, в котором "пройдем" по информации датафреймов каждого региона;
3. По каждому региону в ходе 1000 итераций вложенного цикла:
* выполним 'train_test_split';
* обучим модель линейной регрессии; 
* выполним предсказания на валидационной выборке;
* рассчитаем RMSE модели, запишем в список словаря;
* полученные значения предсказаний передадим функции 'profit', которая рассчитает чистую прибыль и добычу по выборке из ТОП-200 по добыче нефти скважин. Значения запишем в список словаря.

```python
# 1-шаг Предварительная подготовка
# Создадим словари, куда запишем данные rmse, чистой прибыли и добычи по каждому региону
rmse_dict = {0:list(),
             1:list(),
             2:list()}    

profit_dict = {0:list(),
               1:list(),
               2:list()}

total_recovery_dict = {0:list(),
                       1:list(),
                       2:list()}
```

```python
%%time

# для воспроизводимости псевдослучайной последовательности применим np.random.seed(1000) с выборкой из 1000 псевдослучайных чисел
rng = np.random.seed(1000)

# 2-шаг Напишем два цикла, в котором "пройдем" по информации датафреймов каждого региона
for region in range(3):
    
    data = df[df['region'] == region]    
    
    # разделим признаки и целевой признак
    features = data.drop('product', axis=1)
    target   = data['product']
                
    # разделим признаки и целевой признак на обучающий и валидационный набор в пропорции 75:25 c помощью 'train_test_split'
    features_train, features_valid, target_train, target_valid = train_test_split(features,
                                                                                  target,
                                                                                  test_size=0.25,
                                                                                  random_state=12345)
    # объекту model присвоим структуру данных LinearRegression
    model = LinearRegression()
        
    # обучим модель на обучающей выборке
    model.fit(features_train, target_train)
        
# 3-шаг Техникой BOOTSTRAP найдем распределение прибыли и прочих необходимых данных по объектам валидационного набора
    for i in range(1000):
                
        # соберем новую выборку из 500 объектов валидационного набора
        features_valid_sampled = features_valid.sample(n=500,
                                                       replace=True,
                                                       random_state=rng)

        target_valid_sampled = target_valid.sample(n=500,
                                                   replace=True,
                                                   random_state=rng)
        
        target_valid_sampled = pd.Series(target_valid_sampled, 
                                         index=target_valid_sampled.index)

        
        # выполним предсказания на валидационной выборке
        pred = pd.Series(model.predict(features_valid_sampled),
                         index=features_valid_sampled.index)
               
        # рассчитаем RMSE модели и добавим в словарь
        mse = mean_squared_error(target_valid_sampled, pred)
        rmse_dict[region].append(mse ** 0.5)
        
        # передадим функции 'profit' предсказания на валидационной выборке
        net_profit, total_recovery = profit(data,                  # отфильтруем данные по каждому региону
                                            pred,                  # из pred_dict возьмем предсказания по каждому региону
                                            Budget,                # общий бюджет на разработку 200 скважин
                                            Revenue_per_barrel)    # выручка при продаже барреля нефти
        
        # запишем полученные значения в словари 'profit_dict','total_recovery_dict'
        profit_dict[region].append(net_profit)
        total_recovery_dict[region].append(total_recovery)
    
    print('Успех!')
```

* Найдите среднюю прибыль, 95%-й доверительный интервал и риск убытков. Убыток — это отрицательная прибыль.
