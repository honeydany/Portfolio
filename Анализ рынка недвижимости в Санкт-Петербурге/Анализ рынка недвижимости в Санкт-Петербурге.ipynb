## Предобработка данных

### Обработка пропусков

Сначала посчитаем, сколько в таблице пропущенных значений. Для этого достаточно двух методов pandas:

```python
df.isna().sum()
```

**Предлагаемые действия:**

**1. Данные, заполненные системой:**

Данные в столбцах `locality_name`, `cityCenters_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, `days_exposition` были заполнены автоматически системой. 
* Некоторые пропуски, например, в столбцах `parks_nearest`, `ponds_nearest`, `parks_around3000`, `ponds_around3000` - пустые. Возможно, это объекты значительно удалены от парков и водоемов в принципе. Предлагается их пропустить и оставить пустыми;
* Пропуски в `locality_name` (49 шт) можно заменить на "Другой"; 
* Восстановить или заполнить остальные пропуски - `cityCenters_nearest`, `days_exposition` - невозможно, т.к. заполнены системой автоматичеки. Предлагается их пропустить и оставить пустыми;

**2. Ручное заполнение:**

* Данные в столбце `ceiling_height` — имеют данные о высоте потолков (м) в доме. Предлагаем заменить пропуски медианным значением в разрезе населенных пунктов;
* Столбец `floors_total` — всего этажей в доме - не критически важная метрика для выбора квартиры, поэтому пропуски можно заменить на 0;
* Пропуски в столбцах `living_area`(жилая площадь в квадратных метрах (м²)) и `kitchen_area` (площадь кухни в квадратных метрах (м²)) — могут быть связаны с `total_area` — общая площадь;
* Столбцы `is_apartment` — имеет булев тип поэтому пропуски можно заменить на 0;
* Пропуски в `balcony` можно заменить на 0, т.к. балкон в квартире/доме может отсутствовать вовсе;

```python
# заполнение пропусков в столбце ceiling_height
df['locality_name'] = df['locality_name'].fillna('Другой')

# заполнение пропусков в столбце ceiling_height медианным значением в разрезе населенного пункта
df['ceiling_height'] = df.groupby('locality_name')['ceiling_height'].transform(lambda x: x.fillna(x.median()))

# заполнение пропусков в столбце is_apartment, balcony, floors_total
df['is_apartment'] = df['is_apartment'].fillna(0)
df['balcony'] = df['balcony'].fillna(0)
df['floors_total'] = df['floors_total'].fillna(0)

# проверим, все ли пропуски в столбцах заполнены с помощью цикла
colum_names = ['locality_name', 'ceiling_height', 'is_apartment', 'balcony', 'floors_total']

for name in colum_names:
    print('Пропусков в столбце',name,' - ',df[name].isna().sum())
```

```python
# заполненим оставшиеся пропуски в столбце ceiling_height медианным значением совокупности
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())

# проверим
df[df['ceiling_height'].isna()]['ceiling_height'].sum()
```

Пропуски в столбцах `locality_name`, `ceiling_height`, `is_apartment`, `balcony`, `floors_total` заполнены.

Для анализа пропусков в `living_area`, `kitchen_area` рассчитаем коэффициент корреляции между `total_area`-`living_area` и `total_area`-`kitchen_area`:

```python
# рассчитаем коэффициент корреляции между total_area и 'kitchen_area' и 'living_area'
print(df['kitchen_area'].corr(df['total_area']))
print(df['living_area'].corr(df['total_area']))
```

Видим, что в парах `total_area`-`kitchen_area` и `total_area`-`living_area` существует значительная взаимосвязь.

Выполним заполнение пропусков в столбце `living_area` и `kitchen_area` медианным значением в разрезе населенного пункта:

```python
# заполнение пропусков в столбце living_area и kitchen_area медианным значением в разрезе населенного пункта
df['kitchen_area'] = df.groupby('locality_name')['kitchen_area'].transform(lambda x: x.fillna(x.median()))
df['living_area'] = df.groupby('locality_name')['living_area'].transform(lambda x: x.fillna(x.median()))

# проверим
df.isna().sum()
```

Осталось незначительное кол-во пропусков в столбцах, которое не повлияет на дальнейший результат.

### Корретировка типов данных

**Некоторые столбцы имеют некорректный формат данных:**
   * `first_day_exposition` — дата публикации — содержит дату публикации объявления в формате UTC, однако pandas распознал его как тип данных в формате `object` (строка). Решение: необходимо привести данные к типу `datetime`;
   * `floors_total ` — всего этажей в доме — имеет информацию об этажах, однако pandas распознал его как тип данных в формате `float` (числа с плавающей точкой). Решение: необходимо привести данные к типу `int`;
   * `is_apartment` — апартаменты (булев тип), однако pandas распознал его как тип данных в формате `object`. Решение: необходимо привести данные к типу `bool`;
   * `balcony` — число балконов — необходимо привести данные к типу `int`.

```python
# приводим данные к типу time в столбце 'first_day_exposition'
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

# приводим данные к типу int в столбце 'floors_total'
df['floors_total'] = df['floors_total'].astype('int64', errors='ignore')

# приводим данные к типу bool в столбце 'is_apartment'
df['is_apartment'] = df['is_apartment'].astype('bool')

# приводим данные к типу int в столбце 'balcony'
df['balcony'] = df['balcony'].astype('int64', errors='ignore')

# проверим, все ли корректировки выполнены
df.info()
```

#### Вывод

Скорректированы типы данных в столбцах:
* first_day_exposition
* floors_total
* is_apartment
* balcony

## Расчёты и добавление результатов в таблицу

Выполним расчёт некоторых параметров и добавим их в исходный датафрейм, а именно:
* цену квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# расчет цены квадратного метра
df['square_area_price'] = df['last_price']/df['total_area']

# день недели, месяц и год публикации объявления
df['week'] = df['first_day_exposition'].dt.week
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year

# этаж квартиры: варианты — первый, последний, другой добавим в новый столбец 'floor_variant' с помощью функции floor_variant_func

def floor_variant_func(row):

    floor = row['floor']
    floor_max = row['floors_total']
    
    if floor == 1:
        return 'первый'
    if  floor == floor_max:
        return 'последний'
    else:
        return 'другой'

df['floor_variant'] = df.apply(floor_variant_func, axis=1)

# выбороно проведем проверку, правильно ли определились варианты по первым и последним этажам:
print(df.loc[df['floor_variant'] == 'последний', ['floor', 'floors_total', 'floor_variant']])
print(df.loc[df['floor_variant'] == 'первый', ['floor', 'floors_total', 'floor_variant']])

# соотношение жилой и общей площади, а также отношение площади кухни к общей
df['living_total_area'] = df['living_area']/df['total_area']
df['kitchen_total_area'] = df['kitchen_area']/df['total_area']

# проверим, все ли столбцы добавлены
df.columns
```

```python
# в переменную df_describe добавим количественные характеристики по каждому столбцу 
df_describe = df[['total_area','last_price','ceiling_height','days_exposition','rooms']].describe()

# добавим данные по межквартильному размаху данных каждого столбца (добавим "усы" в таблицу)

df_describe.loc['IQR'] = (df_describe.loc['75%'] - df_describe.loc['25%'])*1.5 # межквартильный размах
df_describe.loc['Q1-IQR'] = df_describe.loc['25%'] - df_describe.loc['IQR'] # нижняя граница размаха "усов"
df_describe.loc['Q3+IQR'] = df_describe.loc['75%'] + df_describe.loc['IQR'] # верхняя граница размаха "усов"

# добавим цикл, который определит границы диаграммы размаха с учетом МИН и МАКС значений столбца данных 
for index in ['total_area','last_price','ceiling_height','days_exposition','rooms']:   
    if df_describe.loc['Q1-IQR'][index] < df_describe.loc['min'][index]:
        df_describe.loc['Q1-IQR'][index] = df_describe.loc['min'][index]
        
# добавим цикл, который определит границы диаграммы размаха с учетом МИН и МАКС значений столбца данных 
for index in ['total_area','last_price','ceiling_height','days_exposition','rooms']:   
    if df_describe.loc['Q3+IQR'][index] > df_describe.loc['max'][index]:
        df_describe.loc['Q3+IQR'][index] = df_describe.loc['max'][index]

df_describe
```

В размах «усов» попадают нормальные значения, а за пределами находятся выбросы. 
* Если правый «ус» длиннее максимума, то он заканчивается максимумом. 
* То же — для минимума и левого «уса».
